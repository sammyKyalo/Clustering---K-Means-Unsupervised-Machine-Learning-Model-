```python


# Specify the columns with potential outliers
outlier_columns = ['Annual budget (million USD)',
                   'Program Efficiency results (0% - 100%)',
                   'Number of programs revisions completed in 2022']

# Define a threshold for Z-score to identify outliers (e.g., Z-score > 3 or < -3)
z_threshold = 3

outliers = pd.DataFrame()

for column in outlier_columns:
    z_scores = np.abs((dataframe[column] - dataframe[column].mean()) / dataframe[column].std())
    outlier_mask = z_scores > z_threshold
    outliers = pd.concat([outliers, dataframe[outlier_mask]])

# Extract relevant information from the outliers DataFrame (e.g., program names, countries, regions, differentiations)
outlier_info = outliers[['Program name', 'Country', 'Region', 'Program \ndifferentiation',*outlier_columns]]

# Display the DataFrame containing information about outliers
outlier_info
```

```python

continuous_vars = ['Annual budget (million USD)',
                   'Program Efficiency results (0% - 100%)',
                   'Program Performance results (0% - 100%)',
                   'Number of programs revisions completed in 2022',
                   'Program implementer level of experience (1-12)']

def replace_outliers_with_mean(column):
    outliers_replaced = True
    
    while outliers_replaced:
        mean = column.mean()
        std = column.std()
        
        threshold = 3 * std
        
        outliers_idx = (column - mean > threshold) | (mean - column > threshold)
        if outliers_idx.any():
            column[outliers_idx] = mean
        else:
            outliers_replaced = False
for var in continuous_vars:
    replace_outliers_with_mean(df[var])




```

```python

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


selected_columns = ['Region_Categorical', 'Country_Categorical',
       'Country_Income_Level_Categorical', 'Program_Name_Categorical',
       'Program_Differentiation_Categorical', 'Program_Disease_Categorical',
       'Program_start_year_Categorical',
    'Program implementer level of experience (1-12)',
    'Annual budget (million USD)',
    'Number of programs revisions completed in 2022',
    'Program Efficiency results (0% - 100%)',
    'Program Performance results (0% - 100%)'
]

X = data[selected_columns]

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_std)

explained_variance_ratio = pca.explained_variance_ratio_

plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('PCA - Explained Variance Ratio')
plt.show()

cumulative_explained_variance = explained_variance_ratio.cumsum()
print("Cumulative Explained Variance Ratio:")
print(cumulative_explained_variance)


```

Optimal Number of Clusters

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

k_range = range(1, 11) 

inertias = []


for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

# Create an elbow plot
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertias, 'bo-')
plt.title('Elbow Plot for K-Means Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from kneed import KneeLocator

k_range = range(1, 11) 
inertias = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

# Create an elbow plot
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertias, 'bo-')
plt.title('Elbow Plot for K-Means Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Use KneeLocator to find the optimal number of clusters
kl = KneeLocator(k_range, inertias, curve="convex", direction='decreasing')
optimal_clusters = kl.elbow

print("Optimal number of clusters:", optimal_clusters)
```
